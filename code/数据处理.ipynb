```python
## 表2数据处理
data2 = pd.read_excel('../竞赛发布数据/表2-患者影像信息血肿及水肿的体积及位置.xlsx')
df_data2 = pd.DataFrame(columns=["ID", "流水号"] + list(data2.iloc[:, 2:24].columns) + ["检查次序"])  # 创建一个空的 DataFrame
for i in range(9):
    temp = pd.DataFrame()
    temp = data2.iloc[:, 1+23*i:24+23*i]
    temp.columns = ['流水号'] + list(data2.iloc[:, 2:24].columns)
    # temp.rename(columns={temp.columns[0]: "随访流水号"}, inplace=True)
    temp['检查次序'] = i
    temp = pd.concat([data2["ID"], temp], axis = 1)
    temp = temp.dropna()  # 删除包含任何缺失值的行
    # 使用 concat 方法按行合并  # append在pandas1.4被标记，在2.0中被完全移除
    df_data2 = pd.concat([df_data2, temp], axis = 0, ignore_index=True)
df_data2['流水号'] = df_data2['流水号'].astype('int64')
df_data2.sort_values(by=['ID', '检查次序'], ascending=[True, True], inplace=True)
df_data2.to_csv("../处理后的数据/表2-血肿及水肿的体积及位置.csv", index=False)
df_data2
```

```python
## 表3数据处理

# ED代表水肿，HM代表血肿
data3_ED = pd.read_excel("../竞赛发布数据/表3-患者影像信息血肿及水肿的形状及灰度分布.xlsx", sheet_name="ED")
data3_ED.drop(columns="备注", inplace=True)
data3_ED_columns = ["流水号"]
for i in range(1, len(list(data3_ED.columns))):
    data3_ED_columns.append("ED_" + data3_ED.columns[i])
data3_ED.columns = data3_ED_columns

data3_HM = pd.read_excel("../竞赛发布数据/表3-患者影像信息血肿及水肿的形状及灰度分布.xlsx", sheet_name="Hemo")
data3_HM.drop(columns="备注", inplace=True)
data3_HM_columns = ["流水号"]
for i in range(1, len(list(data3_ED.columns))):
    data3_HM_columns.append("HM_" + data3_HM.columns[i])
data3_HM.columns = data3_HM_columns

data3_ED.to_csv("../处理后的数据/表3_水肿的形状及灰度分布_ED.csv", index=False)
data3_HM.to_csv("../处理后的数据/表3_血肿的形状及灰度分布_HM.csv", index=False)
```

```python
### 附表1-时间点-流水号数据处理

time_data = pd.read_excel("../竞赛发布数据/附表1-检索表格-流水号vs时间.xlsx")  # 流水号、ID、时间对照表
df_time_number = pd.DataFrame(columns=['ID', '重复次数','随访时间点', '流水号', '检查次序'])
for i in range(14):
    temp0 = pd.DataFrame(columns=['随访时间点', '流水号', '检查次序'])
    temp0[["随访时间点", "流水号"]] = time_data.iloc[:, [2+2*i, 2+2*i+1]]
    # 添加一列数字表示第几次检查
    temp0['检查次序'] = i
    temp = pd.concat([time_data[["ID", "重复次数"]], temp0], axis = 1)
    # temp.columns = ["ID", "重复次数", "随访时间点", "随访流水号", "检查次序"]
    temp = temp.dropna()  # 删除包含任何缺失值的行
    df_time_number = pd.concat([df_time_number, temp], axis = 0, ignore_index=True)

df_time_number.to_csv("../处理后的数据/附表1-时间点-流水号.csv", index=False)
df_time_number
```
